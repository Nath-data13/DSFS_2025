```python
# Valeurs manquantes dans la colonne dec_o / attributs o 
manquant_decision = df['dec_o'].isnull().sum() 
print(f"- Dans la colonne 'dec_o' nous avons {manquant_decision} valeurs manquantes") 

# Valeurs manquantes dans la colonne attributs o 
manquant_attributs = df[['attr_o', 'sinc_o','intel_o',	'fun_o','amb_o','shar_o']].isnull().sum() 
print(f"- Nous avons comme valeur manquante dans les 6 colonnes\n{manquant_attributs}") 

# Valeurs manquantes dans la colonne Gender
manquant_genre = df['gender'].isnull().sum()
print(f"- Dans la colonne 'gender' nous avons {manquant_genre} valeurs manquantes") 

# Valeurs manquantes dans la colonne Goal
manquant_objectif = df['goal'].isnull().sum()
pourcentage_obj = np.around((manquant_objectif/(df.shape[0]*df.shape[1]))*100,5)
print(f"- Nous avons {manquant_objectif} valeurs manquantes dans la colonne Goal, ce qui correspond à {pourcentage_obj} %")

# Valeurs manquantes dans la colonne Field_cd
manquant_secteur = df['field_cd'].isnull().sum()
pourcentage_secteur = np.around((manquant_secteur/(df.shape[0]*df.shape[1]))*100,5)
print(f"- Nous avons {manquant_secteur} valeurs manquantes dans la colonne Field_cd, ce qui correspond à {pourcentage_secteur} %")
```

```python
# Certaines colonnes liées aux attributs (attr_o, sinc_o, intel_o, fun_o, amb_o, shar_o) contiennent des valeurs 
# manquantes. Afin d'assurer la continuité de l'analyse, ces valeurs manquantes seront remplacées par la médiane 
# de chaque attribut respectif. Cette méthode permet de préserver la distribution des données et de minimiser 
# l'impact sur les résultats de l'analyse.

# Valeurs manquantes remplacées par la médian
median = df[['attr_o', 'sinc_o','intel_o','fun_o','amb_o','shar_o']].median()
print(f"- La mediane est de :\n{median}")

columns =['attr_o', 'sinc_o','intel_o',	'fun_o','amb_o','shar_o']
df[columns] =df[columns].apply(lambda x: x.fillna(x.median()))
print(f"- Vérification des valeurs manquantes {df[columns].isnull().sum()}")
```

### Quels sont les attributs importants pour un 2éme rendez-vous ? Les attributs les mieux notés pour un 2éme rendez-vous sont la Sincérité, l'Intelligence et l'Humour. Les écarts les plus importants entre une prise de décision d'un 2éme rendez-vous pas, sont l’Attractivité, l’humour et les Intérêts communs. Nous pouvons constater que l’Attractivité, l’Humour et les Interets communs influencent fortement la décision d'un 2eme rendez-vous ou non.


```python
attributes = {
    'attr_o': 'Attractivité',
    'sinc_o': 'Sincérité',
    'intel_o': 'Intelligence',
    'fun_o': 'Humour',
    'amb_o': 'Ambition',
    'shar_o': 'Intérêts communs'
}

notes_melt = pd.DataFrame({
    'Critère': list(attributes.values()),
    '2e rendez-vous': [df[df['dec_o'] == '2ème rendez-vous'][col].mean() for col in attributes],
    'Pas de 2e rendez-vous': [df[df['dec_o'] == 'Pas de 2ème rendez-vous'][col].mean() for col in attributes]
})

notes_melt = notes_melt.melt(id_vars='Critère', var_name='Décision', value_name='Note moyenne')

fig = px.bar(
    notes_melt,
    x='Critère',
    y='Note moyenne',
    color='Décision',
    barmode='group',
    title='Notes moyennes par décision de 2e rendez-vous'
)

fig.update_layout(yaxis_range=[0, 9])
fig.show()
```

### Importance des attributs par genre : Hommes vs Femmes / notes moy par décision de 2e rdv. Ce graphique montre que les hommes et les femmes se retrouvent sur l’importance de l’attractivité et des intérêts communs dans le choix d’un second rendez-vous. En revanche, les femmes semblent plus sensibles à l’humour que les hommes. 

```python

df['gender_HF']=df['gender'].map({0:'Hommes', 1:'Femmes'})

# Calcul des moyennes par décision et par genre
notes_list = []
for genre in ['Hommes', 'Femmes']:
    df_genre = df[df['gender_HF'] == genre]
    for critere_col, critere_nom in attributes.items():
        notes_list.append({
            'Critère': critere_nom,
            'gender_HF': genre,
            '2e rendez-vous': df_genre[df_genre['dec_o'] == '2ème rendez-vous'][critere_col].mean(),
            'Pas de 2e rendez-vous': df_genre[df_genre['dec_o'] == 'Pas de 2ème rendez-vous'][critere_col].mean()
        })

# DataFrame
notes_HF = pd.DataFrame(notes_list)

# Restructuration pour Plotly
notes_melt_HF = notes_HF.melt(
    id_vars=['Critère', 'gender_HF'],
    var_name='Décision',
    value_name='Note moyenne'
)
# Graphique
fig = px.bar(
    notes_melt_HF,
    x='Critère',
    y='Note moyenne',
    color='Décision',
    barmode='group',
    facet_row='gender_HF',
    labels={'gender_HF': 'Genre'},
    title='Notes moyennes par décision de 2e rendez-vous (Hommes vs Femmes)'
)
fig.update_layout(
    yaxis_range=[0, 8],
    width=1000,
    height=800
)
fig.show()
```
